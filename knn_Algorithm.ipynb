```
def dist(x1,x2):
  return np.sqrt(np.sum(x1-x2)**2)
```

```
def knn(X,Y,queryPoint,k=5):
  var=[]
  m=X.shape[0]
  for i in range(m):
    d=dist(queryPoint,X[i])
    var.append((d,Y[i]))
  
  var=sorted(var)
  
  var=var[:k]
  var=np.array(var)
  
  new_vals = np.unique(var[:,1],return_counts=True)
  index=new_vals[1].argmax()
  pred=new_vals[0][index]
  return pred
```

```
queryPoint=np.array([0,0])
```

```
knn(X,Y,queryPoint,6)
```
